```python
# TODO:
# 
# 1. replace column names with meaningful name.
#
# 2. removal of redundant information
#   2.1. redundant columns => "url", "dish_liked", "reviews_list", "menu_item","listed_in(type)","listed_in(city)"
#   2.2. remove duplicate rows
#   2.3. remove records with no price and ratings
# 
# 3. replace null values
#   3.1.    rates -> 0
#   3.2.    phone -> Nil
#   3.3.    rest_type -> replace with Nil
#   3.4.    cuisines -> replace with Nil
#   3.5.    approximate cost -> replace with 0
#
# 4. Handling multiple values:
#   A restaurant has multiple ['cuisine', 'rest_type'] types. Split the multiple items and replace with most occured
#   reducing multiple types for better performance.
```

```python
# replace with meaning column name
df = df.rename(columns={"approx_cost(for two people)": "price"})
df.head(5)
```

```python
df.isna().sum()
```

```python
# split the rate column and convert to float

df['rate'] = df['rate'].replace(['NEW', '-'], None)
df['rate'] = df['rate'].str.split('/').str[0].astype(float)

# Replace NaN with 0
df["rate"] = df["rate"].fillna(0)
```

```python
df.head()
```

```python
# handling multiple rest_types
# replacing with 1st type
df["rest_type"] = df["rest_type"].str.split(",").str[0].str.strip()

# reduced multiple restaurant types
qb = ["Bakery", "Beverage Shop", "Food Court", "Pop Up", "Kiosk", "Food Truck"]
cd = ["Fine Dining", "Mess", "Bhojanalya", "Dhaba", "Lounge"]
bar = ["Bar", "Pub", "Microbrewery", "Club"]
dessert = ["Sweet Shop", "Confectionery", "Dessert Parlor"]

df["rest_type"] = df["rest_type"].replace(qb, "Quick Bites")
df["rest_type"] = df["rest_type"].replace(cd, "Casual Dining")
df["rest_type"] = df["rest_type"].replace(bar, "Bar")
df["rest_type"] = df["rest_type"].replace(dessert, "Dessert")
df["rest_type"] = df["rest_type"].replace(["Takeaway", "Delivery"], "Delivery/Takeaway")

# replace NaN with 'Unknown'
df["rest_type"] = df["rest_type"].fillna("Unknown")
```

```python
df['rest_type'].value_counts()
```

```python
# Handling multiple cuisines

df["cuisines"] = df["cuisines"].str.split(",").str[0].str.strip()

continetal = [
    "Mediterranean",
    "Iranian",
    "Australian",
    "German",
    "Spanish",
    "Middle Eastern",
    "South American",
    "French",
    "Steak",
    "Turkish",
    "Lebanese",
    "Portuguese",
    "African",
    "Italian",
    "Mexican",
    "BBQ",
    "Belgian",
    "British",
    "European",
    "Russian",
    "Arabian",
    "American",
    "Parsi",
]
asian = [
    "Chinese",
    "Indonesian",
    "Singaporean",
    "Sushi",
    "Burmese",
    "Tibetan",
    "Nepalese",
    "Momos",
    "Thai",
    "Japanese",
    "Malaysian",
    "Korean",
    "North Eastern",
    "Vietnamese",
]
north_indian = [
    "Mughlai",
    "Afghan",
    "Kashmiri",
    "Awadhi",
    "Lucknowi",
    "Bengali",
    "Kebab",
    "Rajasthani",
    "Modern Indian",
    "Bohri",
]
south_indian = [
    "Biryani",
    "Charcoal Chicken",
    "Hyderabadi",
    "Chettinad",
    "Andhra",
    "Mangalorean",
    "Maharashtrian",
    "Tamil",
    "Roast Chicken",
    "Kerala",
    "Healthy Food",
    "Salad",
]
east_indian = ["Oriya", "Bihari", "Assamese", "Naga", "Mithai"]
west_indian = ["Maharashtrian", "Goan", "Konkan", "Gujarati"]
fast_food = [
    "Street Food",
    "Rolls",
    "Finger Food",
    "Pizza",
    "Sandwich",
    "Burger",
    "Salad",
    "Bakery",
]
cafe = ["Coffee", "Desserts", "Ice Cream", "Tea", "Bar Food", "Drinks Only", "Juices"]

df["cuisines"] = df["cuisines"].replace(continetal, "Continental")
df["cuisines"] = df["cuisines"].replace(asian, "Asian")
df["cuisines"] = df["cuisines"].replace(north_indian, "North Indian")
df["cuisines"] = df["cuisines"].replace(south_indian, "South Indian")
df["cuisines"] = df["cuisines"].replace(east_indian, "East Indian")
df["cuisines"] = df["cuisines"].replace(west_indian, "West Indian")
df["cuisines"] = df["cuisines"].replace(cafe, "Cafe")
df["cuisines"] = df["cuisines"].replace(fast_food, "Fast Food")

# replace NaN with Unknown
df["cuisines"] = df["cuisines"].fillna("Unknown")
```

```python
df["cuisines"].value_counts()
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
# replace NaN in price

# removing comma from 4,100 
df['price'] = df['price'].str.replace(',','')

# fill NaN with 0
df['price'] = df['price'].fillna(0)

# convert type to int
df['price'] = df['price'].astype(int)
```

```python
df.info()
```

```python
# encoding online_order, book_table to 0,1
df['online_order'] = df['online_order'].map({"Yes":1, "No": 0})
df['book_table'] = df['book_table'].map({"Yes":1, "No": 0})
```
