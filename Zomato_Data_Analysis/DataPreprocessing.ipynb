```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("files/Zomato_Mumbai_Dataset.csv", delimiter='|')
df.head()
```

```python
# basic data informations
df.info()
```

```python
# check redundant records

red_data = df['PAGE NO'] == 'PAGE NO'
red_data
df[red_data]
```

```python
# removing redundant records
df = df[~red_data]
```

```python
# removing unwanted columns

df.drop(["URL", "PAGE NO", "CITY"], axis=1, inplace=True)
df.head()
```

```python
# check null values

df.isna().sum()
```

```python
df[df['CUSINE_CATEGORY'].isna()]
```

```python
# removing the record '15080'

df.drop(labels=15080, axis=0, inplace=True)
```

```python
# verify null once again
df.isna().sum()
```

```python
# replace null values with 'NA'

df.fillna('NA', inplace=True).head()
```

```python
df.isna().sum()
```

```python
df.info()

# from the below information the numerical columns are also in the form of String.
```

```python
# convert numerical column from str to numerical

for name in ['RATING', 'PRICE', 'VOTES']:
    df[name] = pd.to_numeric(df[name], errors='coerce').fillna(0)
```

```python
df.info()
```

```python
df.head()
```

```python
# working with timing column

df['TIMING'].value_counts()
```

```python
# splitting the TIMING column

temp = df["TIMING"].str.split("(", n=1, expand=True)
```

```python
df['TIMING'] = temp[0]
df['DAYS_OPEN'] = temp[1]
```

```python
df.head()
```

```python
df["DAYS_OPEN"] = df["DAYS_OPEN"].str.replace(")", "", regex=False)
```

```python
# check null values in DAYS_OPEN column
df['DAYS_OPEN'].isna().sum()
```

```python
df['DAYS_OPEN'] = df['DAYS_OPEN'].fillna('NA')
```

```python
df.info()
```

```python
# removing restaurant records with rating or votes is 0

useless_data = (df["RATING"] < 1) | (df["VOTES"] < 1)
useless_data
```

```python
df = df[~useless_data]
df.head()
```

```python
# working on rating type column

df['RATING_TYPE'].value_counts()
```

```python
rating_map = {
    # excellent
    "Excellent": "Excellent",
    "Excelente": "Excellent",
    # very good
    "Very Good": "Very Good",
    "Veľmi dobré": "Very Good",
    "Velmi dobré": "Very Good",
    "Bardzo dobrze": "Very Good",
    "Muito Bom": "Very Good",
    "Muy Bueno": "Very Good",
    "Sangat Baik": "Very Good",
    "Çok iyi": "Very Good",
    # good
    "Good": "Good",
    "Bom": "Good",
    "Bueno": "Good",
    "Buono": "Good",
    "İyi": "Good",
    "Baik": "Good",
    "Skvělá volba": "Good",
    "Dobré": "Good",
    "Dobrze": "Good",
    "Skvělé": "Good",
    # average
    "Average": "Average",
    "Ortalama": "Average",
    "Promedio": "Average",
    "Média": "Average",
    "Media": "Average",
    "Priemer": "Average",
    "Průměr": "Average",
    "Biasa": "Average",
    # poor
    "Poor": "Poor",
}

df["RATING_TYPE"] = df["RATING_TYPE"].map(rating_map)
```

```python
df['RATING_TYPE'].value_counts()
```

```python
# working on region column
df.head()
```

```python
# removing unwanted characters from region upto --
## eg. Aureole Hotel-- Andheri East => Andheri East
df["REGION"] = df["REGION"].replace("[a-zA-Z].+--", "", regex=True)
df["REGION"]
```

```python
# removing north, south, east, west


df['REGION'] = df['REGION'].str.lower().replace(r'\b(north|south|east|west)\b', '', regex=True)
df['REGION'].value_counts()
```

```python
# replacing small regions to known region 
df['REGION'] = df['REGION'].str.replace('Bandra Kurla Complex','Bandra',regex=True)
df['REGION'] = df['REGION'].str.replace('CBD-Belapur','CBD Belapur',regex=True)
df['REGION'] = df['REGION'].str.replace('Girgaon Chowpatty','Chowpatty',regex=True)
df['REGION'] = df['REGION'].str.replace('Dadar Shivaji Park','Dadar',regex=True)
df['REGION'] = df['REGION'].str.replace('Flea Bazaar Café|Kamala Mills Compound','Lower Parel',regex=True)
df['REGION'] = df['REGION'].str.replace('Runwal Green','Mulund',regex=True)
df['REGION'] = df['REGION'].str.replace('Mumbai CST Area','Mumbai Central',regex=True)
df['REGION'] = df['REGION'].str.replace('Kopar Khairane|Seawoods|Turbhe|Ulwe','Navi Mumbai',regex=True)
df['REGION'] = df['REGION'].str.replace('New Panvel|Old Panvel','Panvel',regex=True)
df['REGION'] = df['REGION'].str.replace('Kamothe','Sion',regex=True)
df['REGION'] = df['REGION'].str.replace('Ghodbunder Road|Majiwada','Thane',regex=True)
```

```python
# removing duplicate records

df[df.duplicated()]
```

```python
df = df.drop_duplicates()
```

```python
# copying cleaned data into new dataframe
zomato_df = df.copy()
zomato_df.head()
```

```python
# saving the cleaned data into new csv file

import joblib

joblib.dump(zomato_df, 'files/Preprocessed_zomato_dataset.csv')
```
