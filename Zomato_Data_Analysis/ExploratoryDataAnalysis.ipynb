```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import joblib
```

```python
# import processed data file

df = joblib.load('files/Preprocessed_zomato_dataset.csv')
df.head()
```

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = 'vscode'
```

```python
# 1. How many restaurants in mumbai for each type of cuisine?

fig = px.histogram(df, x='CUSINE TYPE', color='CUSINE TYPE',
 title= 'No. of Restaurants by Cuisine Type',
 labels={'CUSINE TYPE':'Cuisine Type'})
fig.show()
```

```python
# 2. percentage of restaurants by rating type

rating_df = df['RATING_TYPE'].value_counts().reset_index()
rating_df
```

```python
rating_percent = df["RATING_TYPE"].value_counts().reset_index()

rating_percent
```

```python
fig = px.pie(
    rating_percent,
    names="RATING_TYPE",
    values="count",
    color="RATING_TYPE",
    title="Percentage of Restaurants by Cuisine",
).update_traces(textposition= 'inside', textinfo = 'percent+label')

fig.show()
```

```python
# 3. which are top high rated seafood restaurants in mumbai?

high_rated = df.sort_values(by="RATING", ascending=False)

top_seafood_restaurants = high_rated[
    high_rated["CUSINE_CATEGORY"].str.contains("seafood", case=False, na=False)
]
top_seafood_restaurants.head(10)
```

```python
# 4. which is the best food truck in mumbai?

best_foodtruck = df[df['CUSINE TYPE'] == 'Food Truck'].sort_values(by= 'RATING', ascending = False)
best_foodtruck.head(3)
```

```python
# Q5) Which places have the highest rated restaurant for each Cuisine Type in Mumbai?

fig = px.histogram(
    high_rated,
    x="REGION",
    y="RATING",
    color="CUSINE TYPE",
    title="High rated restaurant for each cuisine type",
).update_xaxes(categoryorder="total descending")
fig.show()
```

```python
# Q6) What is the Avg Price Distibution of highest rated restaurant for each Cuisine Type in Mumbai?

avg_price = (
    high_rated.groupby(by=["REGION", "CUSINE TYPE"])["PRICE"].mean().reset_index()
)

avg_price
```

```python
px.scatter(
    avg_price,
    x="REGION",
    y="PRICE",
    color='CUSINE TYPE',
    title="Avg price distrubition for each cuisine type",
    symbol="CUSINE TYPE",
).update_traces(marker_size=10).show()
```

```python
# Q9) Is there a relation between Region and Price?


fig = px.scatter(
    df.groupby(["REGION"])["PRICE"].mean().reset_index(),
    x="REGION",
    y="PRICE",
    height=700
).update_traces(marker_size=8)
fig.update_yaxes(range=[0, 2000])
fig.show()
```

```python
# Q10) Find the list of Affordable Restaurants?

# criteria => low price and high rated restaurant
# we'll calculate the economical price (i.e.. less than1/4th of expensive price).

aff_price = df["PRICE"].max() / 4

aff_res = high_rated[high_rated["PRICE"] <= aff_price]
print("total affordable restaurants: ", aff_res["NAME"].count())

aff_res.sort_values(by=["RATING", "PRICE"], ascending=[False, True]).head(5)

```

```python
# logic gap from the below restaurants
# 'Hitchki' has price 1200 and rating 4.9
# but 'Dessertino' has rating of 4.8 and price of 300
# comparing both, 'dessertino' is more affordable and high rated
# the same with the 'The Northern Vibe
#
# To Solve this
#   1. Normalize first
#   2. Include weights for rating and price.

# Normalization
aff_res["rating_norm"] = aff_res["RATING"] / aff_res["RATING"].max()
aff_res["price_norm"] = aff_res["PRICE"] / aff_res["PRICE"].max()

# adding weights (70% to rating and 30% to price)
# dividing by 1000 because rating scale is 0-5 but price is 0-5000.
aff_res["score"] = (0.7 * aff_res["rating_norm"]) + (0.3 * (1- aff_res["price_norm"]))

aff_res.sort_values(by="score", ascending=False)
```

```python
# Q10) Find the list of most Reliable Restaurants?
# The criteria for most Reliable Restaurants would be:-
# 1) Low Price 2) High Rated 3) Large No. of Votes

mean_votes = df['VOTES'].mean()

# Finding list of restaurants that have Votes greater than and equal to Mean of Vote
mean_rest_df = df[['NAME', 'PRICE', 'CUSINE_CATEGORY', 'REGION', 'CUSINE TYPE', 'VOTES']]
mean_rest_df = mean_rest_df[mean_rest_df['VOTES'] > int(mean_votes)]
mean_rest_df.sort_values(by='VOTES', inplace=True)
mean_rest_df
```

```python
reliable_rest_df = pd.merge(
    mean_rest_df, high_rated, how="inner", on=["NAME", "REGION"]
)
reliable_rest_df = reliable_rest_df[
    ["NAME", "PRICE_x", "CUSINE_CATEGORY_x", "REGION", "CUSINE TYPE_x"]
]
reliable_rest_df.rename(
    columns={
        "NAME": "NAME",
        "PRICE_x": "PRICE",
        "CUSINE_CATEGORY_x": "CUSINE_CATEGORY",
        "REGION": "REGION",
        "CUSINE TYPE_x": "CUSINE TYPE",
    },
    inplace=True,
)
reliable_rest_df
```
